```python
# Import pandas, urllib2, requests, and BeautifulSoup modules.
import pandas as pd
import urllib2
import requests
import BeautifulSoup as bs

data_url = "https://github.com/rdhyee/working-open-data/blob/master/data/census/DataDict.txt"
# raw_data_url = "https://raw.github.com/rdhyee/working-open-data/master/data/census/DataDict.txt"
```

```python
# Open data_url.
try:
    response = urllib2.urlopen(data_url)
except urllib2.HTTPError, e:
    print e.code
except urllib2.URLError, e:
    print e.args

# Create url opener as a Mozilla user-agent to avoid blocks.
opener = urllib2.build_opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0')]

response = opener.open(data_url)
read_data = response.read()
```

```python
import nltk

html = urllib2.urlopen(data_url).read()
soup = bs.BeautifulSoup(html)

# Create info_list to keep track of each line of information.
info_list = []

# Go through html to find the DataDict.txt data to
# add to ino_list.
for line in soup.findAll(['pre']):
    for q in line.findAll(['div']):
        info_list.append(nltk.util.clean_html(q.prettify()))

# Take out the first two entries in info_list.
info_list, title_list = info_list[2:], info_list[:2]
# print title_list[0].find('Unit')
# Create a list 'description' to document data description
description = [i for i in title_list[0].split(' ') if i != '']
# Create info_series, a pandas DataFrame version
# of info_list.
# info_series = pd.DataFrame(info_list)
```

```python
data_item_list = []
item_description_list = []
unit_list = []
decimal_list = []
the_rest = []
position_list = []

q = 0
while q < len(info_list):
    # Find 
    position = 0
    if 'ABS' in info_list[q]:
        position = info_list[q].find('ABS')
        unit = 'ABS'
    elif 'PCT' in info_list[q]:
        position = info_list[q].find('PCT')
        unit = 'PCT'
    elif 'MIN' in info_list[q]:
        position = info_list[q].find('MIN')
        unit = 'MIN'
    elif 'DOL' in info_list[q]:
        position = info_list[q].find('DOL')
        unit = 'DOL'
    elif 'AVG' in info_list[q]:
        position = info_list[q].find('AVG')
        unit = 'AVG'
    elif 'TH$' in info_list[q]:
        position = info_list[q].find('TH$')
        unit = 'TH$'
    elif 'SQM' in info_list[q]:
        position = info_list[q].find('SQM')
        unit = 'SQM'
    elif 'RTE' in info_list[q]:
        position = info_list[q].find('RTE')
        unit = 'RTE'
    else:
        pass
    
    item_description_list.append(info_list[q][10:position].rstrip())
    data_item_list.append(info_list[q][:9])
    unit_list.append(unit)
    decimal_list.append(info_list[q][position+4:position+5])
    the_rest.append([i for i in info_list[q][position+5:].split(' ') if i != ''])
    q += 1

us_total_list = []
min_list = []
max_list = []
source_list = []
for i in the_rest:
    us_total_list.append(i[0])
    min_list.append(i[1])
    max_list.append(i[2])
    source_list.append(i[3])
```

```python
data_item_series = pd.Series(data_item_list)
item_description_series = pd.Series(item_description_list)
unit_series = pd.Series(unit_list)
decimal_series = pd.Series(decimal_list)
position_series = pd.Series(unit_list)
us_total_series = pd.Series(us_total_list)
max_series = pd.Series(max_list)
min_series = pd.Series(min_list)
source_series = pd.Series(source_list)

result = {'Data Item':data_item_series,
          'Item Description':item_description_series,
          'Unit':unit_series,
          'Decimal':decimal_series,
          'US Total':us_total_series,
          'Minimum':min_series,
          'Maximum':max_series,
          'Source':source_series
}
result_df = pd.DataFrame(result, columns=['Data Item', 'Item Description', 'Unit', 'Decimal', 'US Total', 'Minimum', 'Maximum', 'Source'])
first_part_df = pd.DataFrame([data_item_series, item_description_series, unit_series, decimal_series],)
the_rest_df = pd.DataFrame(the_rest, columns=description[-4:])


# End result.
result_df
```
