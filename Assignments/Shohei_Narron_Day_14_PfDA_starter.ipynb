## Exercise: calculate how often a given net location appears in frame.u

```python
# FILL IN
import urlparse
import numpy as np
from collections import Counter 

netlocs = []

for link in frame.u.dropna():
    netlocs.append(urlparse.urlparse(link).netloc)

netlocs = pd.Series(Counter(netlocs)).order(ascending=False)






# https://github.com/pydata/pandas/issues/240

assert isinstance(netlocs, Series)

assert set(list(netlocs[:5].iteritems())) == set([(u'www.whitehouse.gov', 169),

     (u'www.monroecounty.gov', 121),

     (u'www.fda.gov', 112),

     (u'www.nasa.gov', 733),

     (u'www.nysdot.gov', 836)])
```

```python
# 2010 is the last available year right now

import os



years = range(1880, 2011)



pieces = []

columns = ['name', 'sex', 'births']



for year in years:

    path = os.path.join(NAMES_DIR, 'yob%d.txt' % year)

    frame = pd.read_csv(path, names=columns)



    frame['year'] = year

    pieces.append(frame)



# Concatenate everything into a single DataFrame

names = pd.concat(pieces, ignore_index=True)
```

```python
names
```

```python
total_births = names.pivot_table('births', rows='year', cols='sex', aggfunc=sum)
```

```python
total_births[:5]
```

```python
# how to calculate the total births / year?
total_births['Per Year'] = total_births['F'] + total_births['M']
total_births[:5]
```

```python
# add prop



def add_prop(group):

    # Integer division floors

    births = group.births.astype(float)



    group['prop'] = births / births.sum()

    return group



names = names.groupby(['year', 'sex']).apply(add_prop)
```

```python
# verify prop

np.allclose(names.groupby(['year', 'sex']).prop.sum(), 1)
```
