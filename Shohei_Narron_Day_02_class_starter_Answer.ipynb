### EXERCISE:  write a function county_fips_for_state that return all county fips for that state fips

```python
# check for CA
county_list = []

# state_fips includes FIPS code for states at this point e.g. '06000', '08000', etc. The following code will exclude them.
def county_fips_for_state(state_fips):
    for each_fips in fips:
        # if the first two numbers (state identifier) match, then append FIPS code to county_list
        if each_fips[:2] == state_fips[:2] and each_fips[2:5] != '000':
            county_list.append(each_fips)
    return sorted(county_list)

# k and fips[k] are FIPS code and county names, respectively.
print [(k, fips[k]) for k in list(county_fips_for_state('06000'))]
```
